```python
'''Proton Loss Function'''

def LsynchrotronHe(E):
    return 2.57E-17 *4**0.79/GF.BetaProton(E)*(E/1E9)**1.28*((E+2E11)/1E9)**-0.2#+4.8*1E-10*xE/E
# Load data from text files
data_files = ['LossFunctionsData/LProtonHelium.txt']
E0_list = []
dEdx_list = []

for file in data_files:
    with open(file, 'r') as f:
        lines = f.readlines()
        for line in lines:
            values = line.split()
    
            E0_list.append(float(values[0]))  # Multiply energy column by 1E6
            dEdx_list.append((float(values[1])+float(values[2])) * 1E-15 + LsynchrotronHe(float(values[0])) )  # Additionnal synchrotron losses at high energy
'''
# Plot the data
plt.plot(E0_list, dEdx_list, 'o', label='Data')

# Set plot labels and title
plt.xlabel('E (eV)')
plt.ylabel('L (eV cm$^{-2}$)')
plt.title('Loss Function of Electrons')

# Set plot scale to logarithmic
plt.xscale('log')
plt.yscale('log')

# Show legend
plt.legend()

# Show the plot
plt.show()
'''
E , LProtonHel = E0_list, dEdx_list


interp1dLProtonHe = interp1d(E, LProtonHel, kind='linear', fill_value='extrapolate')
def LProtonHe(E):
    return interp1dLProtonHe(E)+LsynchrotronHe(E)
#print(LProtonHe(1E10))
```

```python
def Wsinte(x):
    ms = GF.me
    M = GF.me
    
    X = np.logspace(-4, np.log10(x), 1000)
    integ = np.exp(-X**2)

    I = integrate.trapezoid(integ, X, axis=0)
   
    return 2 / np.pi**(1/2) * (I - (1 + ms/M) * x * np.exp(-x**2))

def Wsint(x):
    ms = GF.me
    M = GF.mp
    
    X = np.logspace(-4, np.log10(x), 1000)
    integ = np.exp(-X**2)

    I = integrate.trapezoid(integ, X, axis=0)
   
    return 2 / np.pi**(1/2) * (I - (1 + ms/M) * x * np.exp(-x**2))

El=np.logspace(-2,10,500)
WPinterpl=Wsint(El)
WEinterpl=Wsinte(El)


WPinterp=interp1d(El,WPinterpl,fill_value='extrapolate')
WEinterp=interp1d(El,WEinterpl,fill_value='extrapolate')
def WsE(x):
    return WEinterp(x)
def WsP(x):
    return WPinterp(x)
```

```python

    
def LCOULOMBElectron(E,DensityH2,DensityH,DensityHe,DensityE,Te): # Mannheim 1994
    

    ionfrac=DensityE/(DensityH2+DensityH+DensityHe) 
    if type(Te)==np.ndarray:
        ones_array = np.ones(Te.shape)
    else:
        ones_array = 1
    beta=ones_array*E
    betae=GF.BetaElectron(GF.kEneV*Te)
 
    L=30*GF.c*GF.sigmaT*GF.me*GF.c**2/beta *ionfrac*WsE(beta/betae)*1/(beta*GF.c)  #Reinhard Schlickeiser eq 5.3.34

    return L*6.242*1E11

def LCOULOMBProton(E,DensityH2,DensityH,DensityHe,DensityE,Te): # Mannheim 1994
    

    ionfrac=DensityE/(DensityH2+DensityH+DensityHe) 
    if type(Te)==float:
        ones_array = np.ones(1)
    elif type(Te)==np.ndarray:
        ones_array = np.ones(Te.shape)
    beta=ones_array*E
    betae=GF.BetaProton(GF.kEneV*Te)
 
    L=30*GF.c*GF.sigmaT*GF.me*GF.c**2/beta *ionfrac*WsP(beta/betae)*1/(beta*GF.c)  #Reinhard Schlickeiser eq 5.3.34

    return L*6.242*1E11

def LElectron(E,DensityH2,DensityH,DensityHe,DensityE,Te):
    #fH=sum(diffS_array*DensityH/(DensityH2+DensityH+DensityHe+DensityE))
    return LElectronH(E)#+LElectronH2(E)+LElectronHe(E)+LCOULOMBElectron(E,DensityH2,DensityH,DensityHe,DensityE,Te)

def LProton(E,DensityH2,DensityH,DensityHe,DensityE,Te):
    
    return LProtonH(E)+LProtonH2(E)+LProtonHe(E)+LCOULOMBProton(E,DensityH2,DensityH,DensityHe,DensityE,Te)


```
